```python
from nltk.corpus import webtext
text=webtext.raw("firefox.txt")
print(text)
```

## Type VS Token

### Type is every unique word in text
### Token is every count of word in text

```python
from nltk.tokenize import word_tokenize
tokens=word_tokenize(text)
types=set(tokens)
```

there 96066 word in text

```python
len(tokens)
```

there 9328 unique word in text

```python
len(types)
```
